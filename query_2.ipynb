```python
neighbourhoods_collection = db["neighbourhoods"]
listings_collection = db["listings_main"]


pipeline = [
    {
        '$lookup': {
            'from': 'listings_main',
            'localField': 'neighbourhood',
            'foreignField': 'neighbourhood',
            'as': 'matchingListings'
        }
    },
    {
        '$match': {
            'matchingListings': {
                '$size': 0
            }
        }
    },
    {
        '$group': {
            '_id': '$neighbourhood'
        }
    },
    {
        '$project': {
            '_id': 0,
            'neighbourhood': '$_id'
        }
    }
]

# Execute the aggregation pipeline
result = neighbourhoods_collection.aggregate(pipeline)

# Display the result
for doc in result:
    print(doc["neighbourhood"])
```
